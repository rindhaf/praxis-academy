```python
import matplotlib.pyplot as plt
cf =  pd.read_csv('cafe.csv')
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.hist(cf['Type'], bins = 5)

plt.title('Type Distribution')
plt.xlabel('Type')
plt.ylabel('#Type')
plt.show()
```

```python
import matplotlib.pyplot as plt
g = plt.figure()
ax = g.add_subplot(1,1,1)
ax.scatter(cf['Total'], cf['Type'])

plt.title('Total and Type distribution')
plt.xlabel('Total')
plt.ylabel('Type')
plt.show()
```
