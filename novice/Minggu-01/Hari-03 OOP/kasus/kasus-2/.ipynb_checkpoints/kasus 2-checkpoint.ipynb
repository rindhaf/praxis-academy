```python
##python2 program showing of sqrt()
import math

def check(n):
    if n == 1:
        return False
    
    for x in range(2, (int)(math.sqrt(n))+1):
        if n % x == 0:
            return False
        return True
    
n = 23
if check(n):
    print("the number is prime")
else:
    print("the number is not prime")

print(n)
```

```python
##python2 program showing graphical representation of tan() function

import math
import numpy as np
import matplotlib.pyplot as plt

in_array = np.linspace(0, np.pi, 10)

out_array = []

for i in range(len(in_array)):
    out_array.append(math.tan(in_array[i]))
    i += 1
    
print("in_array : ", in_array)  
print("\nout_array : ", out_array)  
  
# red for numpy.sin()  
plt.plot(in_array, out_array, color = 'red', marker = "o")  
plt.title("math.tan()")  
plt.xlabel("X")  
plt.ylabel("Y")  
plt.show()
```

```python
##python2 program showing graphical representation of cos() function

import math
import numpy as np
import matplotlib.pyplot as plt

in_array = np.linspace(-(3 * np.pi), 3 * np.pi, 30)

out_array = []

for i in range(len(in_array)): 
    out_array.append(math.cos(in_array[i])) 
    i += 1
  
   
print("in_array : ", in_array)  
print("\nout_array : ", out_array)  

plt.plot(in_array, out_array, color = 'blue', marker = "o")  
plt.title("math.cos()")  
plt.xlabel("X")  
plt.ylabel("Y")  
plt.show()  
```

```python
##python2 program showing graphical representation of sin() function
import math 
import matplotlib.pyplot as plt  
  
in_array = [-3.14159265, -2.57039399, -0.28559933, 
            0.28559933, 2.57039399,  3.14159265] 
  
out_array = [] 
  
for i in range(len(in_array)): 
    out_array.append(math.sin(in_array[i])) 
    i += 1
  
   
print("in_array : ", in_array)  
print("\nout_array : ", out_array)  
  
plt.plot(in_array, out_array, color = 'green', marker = "o")  
plt.title("math.sin()")  
plt.xlabel("X")  
plt.ylabel("Y")  
plt.show()  
```
